```python
attribute = 'date'
print('attribute name:',attribute)
print('unique nunmbers' ,len(myData[attribute].unique()))
print('unique nunmbers propotion',len(myData[attribute].unique())/myData.shape[0] )      
print('null numbers', myData.shape[0] - myData[attribute].count())
print('0 value numbers', myData[myData[attribute] == 0].shape[0] )
```

```python
myData[attribute].value_counts().head(20).plot(kind='bar')
plt.savefig('hist.png')
#plt.yscale('log')
```

```python
def str_to_datetime (x):
    return dt.datetime.strptime(x, '%Y%m%d')
myData['date'] = myData['date'].astype(str)
myData['date'] = myData['date'].apply(str_to_datetime)
```

```python
myData['date'].dtypes
```

```python
numrecords = len(myData)
sns.set(font_scale=1.2)
myData.assign(trx = np.ones(numrecords)).set_index(myData['date']).resample('w')\
    .count().trx.plot(title = 'Weekly Transactions')
plt.savefig('hist.png')
```
